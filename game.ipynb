```python
import random
player=None
choices=['rock','scissor','paper']
computer=random.choice(choices)
print(computer)
while player not in choices:
  player=input("ENTER A VALUE : ROCK SCISSOR PAPER ").lower()
  if computer==player:
    print(f"COMPUTER :{computer}")
    print(f"PLAYER :{player}")
    print("Tie !!")
  elif player=='rock':
    if computer=="paper":
      print("computer win !")
    if computer=="scissor":
      print("player win !")

  elif player=="paper":
    if computer=="rock":
      print("player win")
    if computer=="scissor":
      print("computer win")

  elif player=="scissor":
    if computer=="paper":
      print("player win")
    if computer=="rock":
      print("computer win")
```
